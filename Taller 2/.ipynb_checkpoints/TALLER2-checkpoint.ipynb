El propósito de este taller es demostrar el afianzamiento de las habilidades requeridas para analizar datos estadísticos a partir de una o dos variables con respecto a los resultados. Para cumplir dicho objetivo se buscó un dataset de Kaggle, siendo ese dataset los resultados de los casos de COvID en una muestra de 100 personas en Estados Unidos.

```python
#Aqui se importa la libreria pandas y se lee el archivo csv
import pandas as pd

file_path ="C:/Users/accc_/ICESI\Analisis exploratorio/Tareas-Analisis-Exploratorio/Tareas-Analisis-Exploratorio/Taller 2/covid_toy.csv"

#Aqui se vuelve a leer el archivo csv
df = pd.read_csv(file_path)

# Se muestra el dataframe del archivo csv
df
```

# DESPÚES DE VER EN PRIMERA INSTANCIA LA TABLA, SE CONCLUYO QUE LAS VARIABLES MÁS RELEVANTES EN EL ESTUDIO SON LA FIEBRE Y LA TOS 
Por eso se determinó que lo que debería hacerse es ver la relación entre ambos síntomas y el COvID, tratando de encontrar la probabilidad de que las personas con ambos presenten el virus.

```python
# Agrupar los datos por fiebre y tos y contar el número de casos de COVID-19
grouped = df.groupby(['fever', 'cough', 'has_covid']).size().unstack(fill_value=0)

# Calcular la probabilidad de tener COVID-19 según los síntomas
grouped['prob_covid'] = grouped['Yes'] / (grouped['Yes'] + grouped['No'])
```

# Habiendo agrupado las variables en distintas tablas con tipos de valores similares (contando los casos en que la persona tenga CoVID y los síntomas), se procede a realizar tablas con los datos. Las tablas escogidas fueron las barras apiladas y el mapa de calor. 

Las barras apiladas se escogieron porque muestran los dos sintomas en las diferentes combinaciones posibles y cuantos resultan en casos positivos de CoVid.

 El mapa de calor se escogió porque también muestra las combinaciones pero en una forma más intuitiva, demostrando que la fiebre no es la causa determinante sino está acompañada de una tos.

```python
import matplotlib.pyplot as plt

# Ejemplo simplificado de datos agrupados
grouped['Total'] = grouped['Yes'] + grouped['No']
grouped['No'] = grouped['No'] / grouped['Total']
grouped['Yes'] = grouped['Yes'] / grouped['Total']

# Crear el gráfico de barras apiladas
grouped[['No', 'Yes']].plot(kind='bar', stacked=True, color=['red', 'green'])

plt.title('Probabilidad de COVID-19 según Síntomas')
plt.xlabel('Combinación de Síntomas')
plt.ylabel('Proporción')
plt.legend(['No COVID', 'Yes COVID'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame solo con las probabilidades
prob_data = grouped['prob_covid'].unstack()

# Crear el heatmap
sns.heatmap(prob_data, annot=True, cmap="YlGnBu")

plt.title('Mapa de Calor de la Probabilidad de COVID-19')
plt.xlabel('Tos')
plt.ylabel('Fiebre')
plt.show()
```
